```python
import os
from pathlib import Path
from tqdm import tqdm

def auto_annotate_images(dataset_path, model_path, img_size=800):
    # Initialize YOLO model
    model = YOLO(model_path)

    # Iterate through train, test, and valid directories
    for split in ['train', 'test', 'valid']:
        images_dir = Path(dataset_path) / split / 'images'
        labels_dir = Path(dataset_path) / split / 'labels'

        # Ensure labels directory exists
        labels_dir.mkdir(parents=True, exist_ok=True)

        # Iterate through images and generate annotations
        for image_path in tqdm(list(images_dir.glob("*.jpg"))):  # Update extension if not .jpg
            # Predict with YOLOv8
            results = model(image_path, imgsz=img_size, conf=0.25)

            # Save results in YOLO format
            for result in results:
                label_file = labels_dir / (image_path.stem + '.txt')
                with open(label_file, 'w') as f:
                    for box in result.boxes:
                        cls, *xywh = box.xywhn[0].tolist()  # Normalized YOLO format: cls, x, y, w, h
                        f.write(f"{int(cls)} {' '.join(map(str, xywh))}\n")

if __name__ == "__main__":
    dataset_path = "/content/dataset"  # Path to your dataset
    model_path = "yolov8s.pt"         # Path to YOLOv8 model

    auto_annotate_images(dataset_path, model_path)
```
